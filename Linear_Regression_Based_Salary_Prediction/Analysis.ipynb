```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("Linear_Regression_Salary_Dataset.csv")
df.drop([5878,0],inplace=True)
plt.clf()
array=df['Education Level'].value_counts()
print(array)
plt.pie(array,labels=array.index)
plt.show()

# Priority Adding

# Create a mapping dictionary for salary priority
priority_mapping = {
    'High School': 1,
    'Bachelor\'s': 2,
    'Master\'s': 3.5,
    "Bachelor's Degree":3,
    "Master's Degree":4,
    'PhD': 5
}

df['Salary Priority'] = df['Education Level'].map(priority_mapping)
```
