```python
import sympy as sp
import numpy as np
```

```python
x = sp.symbols('x')
print(x+1)
```

```python
# Expresssion transformations 
display(x+1)
display((x**2 - x).factor())
display(sp.expand((x*(x-1))))
```

```python
# Common Mathematical functions
ratnum1 = sp.Rational(5,3) # Stores in P/Q format
display(ratnum1)
display(ratnum1.evalf())
display(ratnum1.evalf(5))
display(sp.sin(sp.pi/6))
display(sp.exp(10))
```

```python
# Expression Substitutions
expression = x + 8
answer = expression.subs(x,3)
answer

y = sp.symbols('y') # symbolizing y
expression = x**2 + 2*y*x + 8
answer = expression.subs([(x,3),(y,1)])
answer

# Using other modules - using lambdify
domain = np.linspace(-3,3,7)
expr = sp.sin(x)
f = sp.lambdify(x,expr,'numpy')
range = f(domain)
display(range)

# Simplifying Equations
expr = ((x**3 + 2*x - x**2 - x - 1)/(x**2 + 2*x + 1))
display(sp.simplify(expr))

expr = ((x**3 + 2*x - x**2 - 5*x - 1))
# collect - ordering in the descending degree of certain variable
display(expr.collect(x))

# cancel - simplifies any expression to rational form and also cuts off common multiplied terms in numerator and denominator
display(sp.cancel(((x - 3) * (x + 2) * (x - 4)) / ((x - 1)*(x - 3))))

# apart - partial fractionization of an expression
display(sp.apart((1 / (x + 3)) + (x / (x**2 - 3)) + ((x - 1) / (x**2 + 3))))

# Trigonometric helpful functions
# trigsimp() Simplifies trigonometric expressions
# expand_trig() for expanding trig expressions

# Powers functions
# powsimp() Symplifies power expressions
# expand_power_exp() and expand_power_base() are also used

# Logarithmic functions
# expand_log(), logcombine()

# Binomial expansions can be computed using binomial() and evalf() functions
```
